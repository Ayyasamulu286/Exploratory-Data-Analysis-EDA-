# Task 5: Exploratory Data Analysis (EDA)

**Data Analyst Internship — DataX Labs**

Objective: Extract insights using visual and statistical exploration.

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.dpi'] = 120
print('Libraries imported successfully!')
```

## 1. Data Loading and Inspection

```python
# Load the dataset
df = pd.read_csv('Mall_Customers.csv')
print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
df.head(10)
```

```python
# Dataset Info
print('=== Dataset Info ===')
df.info()
print('\n=== Missing Values ===')
print(df.isnull().sum())
```

```python
# Statistical Summary
print('=== Statistical Summary ===')
df.describe()
```

```python
# Categorical Analysis
print('=== Gender Distribution ===')
print(df['Gender'].value_counts())
print(df['Gender'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
```

## 2. Univariate Analysis
### 2.1 Histograms — Distribution of Numerical Features

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Distribution of Numerical Features', fontsize=16, fontweight='bold', y=1.02)
cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
colors = ['#4472C4', '#ED7D31', '#70AD47']

for i, (col, color) in enumerate(zip(cols, colors)):
    axes[i].hist(df[col], bins=20, color=color, edgecolor='white', alpha=0.85)
    axes[i].set_title(f'Distribution of {col}', fontsize=12)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    mean_val = df[col].mean()
    axes[i].axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}')
    axes[i].legend()

plt.tight_layout()
plt.savefig('histogram.png', dpi=150, bbox_inches='tight')
plt.show()

# Observations
print('Observations:')
print('- Age: Slightly right-skewed. Peak around age 30. Most customers aged 20-40.')
print('- Annual Income: ~Normal distribution centered at $60k. Few outliers above $120k.')
print('- Spending Score: Nearly uniform with slight peaks at extremes.')
```

### 2.2 Gender Distribution — Pie Chart

```python
fig, ax = plt.subplots(figsize=(6, 6))
gender_counts = df['Gender'].value_counts()
ax.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%',
       colors=['#ED7D31', '#4472C4'], startangle=90)
ax.set_title('Gender Distribution', fontsize=14, fontweight='bold')
plt.show()
print('Observation: 56% Female, 44% Male — slightly female-dominant dataset.')
```

## 3. Bivariate Analysis
### 3.1 Boxplots — Outlier Detection by Gender

```python
palette = {'Male': '#4472C4', 'Female': '#ED7D31'}
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Boxplot — Outlier Detection by Gender', fontsize=16, fontweight='bold')

for i, col in enumerate(cols):
    sns.boxplot(data=df, x='Gender', y=col, ax=axes[i], 
                hue='Gender', palette=palette, legend=False)
    axes[i].set_title(f'{col} by Gender')

plt.tight_layout()
plt.savefig('boxplot.png', dpi=150, bbox_inches='tight')
plt.show()

print('Observations:')
print('- Age: Similar distributions for both genders.')
print('- Income: Males slightly higher median income than females.')
print('- Spending Score: Females have marginally higher median spending score.')
```

### 3.2 Scatter Plots — Relationship Between Features

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Scatter Plots — Relationship Analysis', fontsize=16, fontweight='bold')

sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)',
                hue='Gender', palette=palette, ax=axes[0], alpha=0.7, s=60)
axes[0].set_title('Age vs Spending Score')

sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Gender', palette=palette, ax=axes[1], alpha=0.7, s=60)
axes[1].set_title('Annual Income vs Spending Score')

plt.tight_layout()
plt.savefig('scatterplot.png', dpi=150, bbox_inches='tight')
plt.show()

print('Observations:')
print('- Age vs Spending Score: Negative trend. Younger customers spend more.')
print('- Income vs Spending Score: No clear linear trend. Visible clusters emerge.')
print('- 5 potential customer segments visible in the income-spending scatter.')
```

## 4. Multivariate Analysis
### 4.1 Correlation Heatmap

```python
corr = df[cols].corr()
print('Correlation Matrix:')
print(corr)

fig, ax = plt.subplots(figsize=(8, 6))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.3f', cmap='RdYlGn', ax=ax, mask=mask,
            linewidths=0.5, vmin=-1, vmax=1, center=0, square=True,
            annot_kws={'size': 14, 'weight': 'bold'})
ax.set_title('Correlation Heatmap', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

print('Observations:')
print('- Age vs Spending Score: -0.327 (moderate negative correlation)')
print('- Age vs Income: -0.012 (virtually no correlation)')
print('- Income vs Spending Score: 0.010 (virtually no correlation)')
print('- No multicollinearity detected among features.')
```

### 4.2 Pairplot — All Pairwise Relationships

```python
g = sns.pairplot(df[cols + ['Gender']], hue='Gender', palette=palette,
                 diag_kind='kde', plot_kws={'alpha': 0.6, 's': 40})
g.fig.suptitle('Pairplot — Pairwise Relationships by Gender', y=1.02,
               fontsize=14, fontweight='bold')
plt.savefig('pairplot.png', dpi=150, bbox_inches='tight')
plt.show()

print('Observations:')
print('- KDE diagonal confirms individual distributions.')
print('- Age-Spending negative trend confirmed across both genders.')
print('- Income-Spending shows clear cluster formations.')
print('- Females cluster slightly higher on spending scores.')
```

## 5. Key Findings Summary

```python
print('=== KEY FINDINGS SUMMARY ===')
findings = [
    '1. Most customers are aged 20-40 years (right-skewed age distribution).',
    '2. Age negatively correlates with spending score (r = -0.327). Younger = higher spenders.',
    '3. Annual income does NOT predict spending behavior (r = 0.010).',
    '4. Five natural customer clusters visible in Income vs Spending scatter plot.',
    '5. Females (56%) slightly outnumber males and show marginally higher spending scores.',
    '6. A few high-income outliers (>$120k) exist but do not dominate spending behavior.',
    '7. Dataset is clean — 0 missing values, all features in valid ranges.',
]
for f in findings:
    print(f)
```

## 6. Conclusion

EDA on the Mall Customers dataset revealed that **age is the strongest predictor of spending behavior**, while annual income alone is insufficient for predicting customer spending. The data clearly shows segmentation potential — 5 distinct customer clusters are visible in scatter plots, making this dataset an excellent candidate for K-Means Clustering analysis. EDA is a critical foundation step before any predictive modeling or business decision-making.
